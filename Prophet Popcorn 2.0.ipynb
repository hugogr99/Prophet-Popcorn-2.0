```python
import pandas as pd
import json
from collections import Counter
```

```python
# Abre os databases
dados_filme = pd.read_csv('tmdb_5000_movies.csv')
dados_creditos = pd.read_csv('tmdb_5000_credits.csv')
```

```python
# Seleciona as variáveis de interesse para o classificador (média de nota, elenco, equipe e número de votos)
titulo = dados_filme['title']
nota = dados_filme['vote_average']
elenco = dados_creditos['cast']
equipe = dados_creditos['crew']
votos = dados_filme['vote_count']

# Junta as colunas de interesse.
dados = pd.concat([titulo,elenco,equipe,nota,votos], axis=1)

# Seleciona somente os filmes que tiveram mais de 100 votos para computar a média. 
dados = dados[dados['vote_count']>=100]

# Os membros da equipe que serão levados em consideração são: Diretor, Produtor e Roteirista.
dados = pd.concat([dados['title'],dados['cast'],dados['crew'],dados['vote_average']],axis=1)
dados.columns = ['Título','Elenco','Equipe','Nota']

#Cria as categorias de nota que serão utilizadas no classificador.
filmes_3 = dados.loc[(dados.Nota<4)]
filmes_4 = dados.loc[(dados.Nota>=4) & (dados.Nota<5)]
filmes_5 = dados.loc[(dados.Nota>=5) & (dados.Nota<6)]
filmes_6 = dados.loc[(dados.Nota>=6) & (dados.Nota<7)]
filmes_7 = dados.loc[(dados.Nota>=7) & (dados.Nota<8)]
filmes_8 = dados.loc[(dados.Nota>8)]
```

```python
## Separa o elenco e a equipe de cada filme, de acordo com a nota que o filme recebeu. 
## Cada categoria de nota possui uma lista de equipe e de elenco, com seus integrantes.

f3_elenco = []
f3_equipe = []

for elenco in filmes_3['Elenco']:
    elenco = json.loads(elenco)
    for membro in elenco:
        f3_elenco.append(membro['name'])

for equipe in filmes_3['Equipe']:
    equipe = json.loads(equipe)
    for membro in equipe:
        if membro['job'] == 'Director':
            f3_equipe.append(membro['name'])
        if membro['job'] == 'Screenplay':
            f3_equipe.append(membro['name'])
        if membro['job'] == 'Producer':
            f3_equipe.append(membro['name'])
        
        
f4_elenco = []
f4_equipe = []

for elenco in filmes_4['Elenco']:
    elenco = json.loads(elenco)
    for membro in elenco:
        f4_elenco.append(membro['name'])

for equipe in filmes_4['Equipe']:
    equipe = json.loads(equipe)
    for membro in equipe:
        if membro['job'] == 'Director':
            f4_equipe.append(membro['name'])
        if membro['job'] == 'Screenplay':
            f4_equipe.append(membro['name'])
        if membro['job'] == 'Producer':
            f4_equipe.append(membro['name'])
            

f5_elenco = []
f5_equipe = []

for elenco in filmes_5['Elenco']:
    elenco = json.loads(elenco)
    for membro in elenco:
        f5_elenco.append(membro['name'])

for equipe in filmes_5['Equipe']:
    equipe = json.loads(equipe)
    for membro in equipe:
        if membro['job'] == 'Director':
            f5_equipe.append(membro['name'])
        if membro['job'] == 'Screenplay':
            f5_equipe.append(membro['name'])
        if membro['job'] == 'Producer':
            f5_equipe.append(membro['name'])
            

f6_elenco = []
f6_equipe = []

for elenco in filmes_6['Elenco']:
    elenco = json.loads(elenco)
    for membro in elenco:
        f6_elenco.append(membro['name'])

for equipe in filmes_6['Equipe']:
    equipe = json.loads(equipe)
    for membro in equipe:
        if membro['job'] == 'Director':
            f6_equipe.append(membro['name'])
        if membro['job'] == 'Screenplay':
            f6_equipe.append(membro['name'])
        if membro['job'] == 'Producer':
            f6_equipe.append(membro['name'])
            
            
f7_elenco = []
f7_equipe = []

for elenco in filmes_7['Elenco']:
    elenco = json.loads(elenco)
    for membro in elenco:
        f7_elenco.append(membro['name'])

for equipe in filmes_7['Equipe']:
    equipe = json.loads(equipe)
    for membro in equipe:
        if membro['job'] == 'Director':
            f7_equipe.append(membro['name'])
        if membro['job'] == 'Screenplay':
            f7_equipe.append(membro['name'])
        if membro['job'] == 'Producer':
            f7_equipe.append(membro['name'])

            
f8_elenco = []
f8_equipe = []

for elenco in filmes_8['Elenco']:
    elenco = json.loads(elenco)
    for membro in elenco:
        f8_elenco.append(membro['name'])

for equipe in filmes_8['Equipe']:
    equipe = json.loads(equipe)
    for membro in equipe:
        if membro['job'] == 'Director':
            f8_equipe.append(membro['name'])
        if membro['job'] == 'Screenplay':
            f8_equipe.append(membro['name'])
        if membro['job'] == 'Producer':
            f8_equipe.append(membro['name'])


# Conta quantas vezes cada nome aparece em cada categoria de filme. Exemplo: Leonardo DiCaprio aparece 12 vezes em filmes
# com notas >7 e <8.
# Cada categoria tem sua recorrencia de atores. 
# Além disso, a lista atores_elenco/membros_elenco guardam os nomes dos atores e dos membros do elenco sem repetição.

atores_elenco_3 = []
membros_equipe_3 = []
f3_elenco_recorrencia = dict(Counter(f3_elenco))
f3_equipe_recorrencia = dict(Counter(f3_equipe))
for membro in f3_elenco:
    if membro not in atores_elenco_3:
        atores_elenco_3.append(membro)
for membro in f3_equipe:
    if membro not in membros_equipe_3:
        membros_equipe_3.append(membro)

atores_elenco_4 = []
membros_equipe_4 = []
f4_elenco_recorrencia = dict(Counter(f4_elenco))
f4_equipe_recorrencia = dict(Counter(f4_equipe))
for membro in f4_elenco:
    if membro not in atores_elenco_4:
        atores_elenco_4.append(membro)
for membro in f4_equipe:
    if membro not in membros_equipe_4:
        membros_equipe_4.append(membro)

atores_elenco_5 = []
membros_equipe_5 = []
f5_elenco_recorrencia = dict(Counter(f5_elenco))
f5_equipe_recorrencia = dict(Counter(f5_equipe))
for membro in f5_elenco:
    if membro not in atores_elenco_5:
        atores_elenco_5.append(membro)
for membro in f5_equipe:
    if membro not in membros_equipe_5:
        membros_equipe_5.append(membro)

atores_elenco_6 = []
membros_equipe_6 = []
f6_elenco_recorrencia = dict(Counter(f6_elenco))
f6_equipe_recorrencia = dict(Counter(f6_equipe))
for membro in f6_elenco:
    if membro not in atores_elenco_6:
        atores_elenco_6.append(membro)
for membro in f6_equipe:
    if membro not in membros_equipe_6:
        membros_equipe_6.append(membro)

atores_elenco_7 = []
membros_equipe_7 = []
f7_elenco_recorrencia = dict(Counter(f7_elenco))
f7_equipe_recorrencia = dict(Counter(f7_equipe))
for membro in f7_elenco:
    if membro not in atores_elenco_7:
        atores_elenco_7.append(membro)
for membro in f7_equipe:
    if membro not in membros_equipe_7:
        membros_equipe_7.append(membro)


atores_elenco_8 = []
membros_equipe_8 = []
f8_elenco_recorrencia = dict(Counter(f8_elenco))
f8_equipe_recorrencia = dict(Counter(f8_equipe))
for membro in f8_elenco:
    if membro not in atores_elenco_8:
        atores_elenco_8.append(membro)
for membro in f8_equipe:
    if membro not in membros_equipe_8:
        membros_equipe_8.append(membro)

# Calcula o número de possibilidades de membro de equipe.
total_equipe = len(membros_equipe_3+membros_equipe_4+membros_equipe_5+membros_equipe_6+membros_equipe_7+membros_equipe_8)
# Calcula o número de possibilidades de atores.
total_elenco = len(atores_elenco_3+atores_elenco_4+atores_elenco_5+atores_elenco_6+atores_elenco_7+atores_elenco_8)
```

```python
# Testa qual a probabilidade de cada ator/membro da equipe aparecer na produção de filmes de cada categoria.
dic_elenco_3 = {}
dic_equipe_3 = {}
for nome in f3_elenco:
    p = (f3_elenco_recorrencia[nome]+1)/(total_elenco)
    dic_elenco_3[nome]=p
for nome in f3_equipe:
    p = (f3_equipe_recorrencia[nome]+1)/(total_equipe)
    dic_equipe_3[nome]=p
    
dic_elenco_4 = {}
dic_equipe_4 = {}
for nome in f4_elenco:
    p = (f4_elenco_recorrencia[nome]+1)/(total_elenco)
    dic_elenco_4[nome]=p
for nome in f4_equipe:
    p = (f4_equipe_recorrencia[nome]+1)/(total_equipe)
    dic_equipe_4[nome]=p
    
dic_elenco_5 = {}
dic_equipe_5 = {}
for nome in f5_elenco:
    p = (f5_elenco_recorrencia[nome]+1)/(total_elenco)
    dic_elenco_5[nome]=p
for nome in f5_equipe:
    p = (f5_equipe_recorrencia[nome]+1)/(total_equipe)
    dic_equipe_5[nome]=p
    
dic_elenco_6 = {}
dic_equipe_6 = {}
for nome in f6_elenco:
    p = (f6_elenco_recorrencia[nome]+1)/(total_elenco)
    dic_elenco_6[nome]=p
for nome in f6_equipe:
    p = (f6_equipe_recorrencia[nome]+1)/(total_equipe)
    dic_equipe_6[nome]=p
    
dic_elenco_7 = {}
dic_equipe_7 = {}
for nome in f7_elenco:
    p = (f7_elenco_recorrencia[nome]+1)/(total_elenco)
    dic_elenco_7[nome]=p
for nome in f7_equipe:
    p = (f7_equipe_recorrencia[nome]+1)/(total_equipe)
    dic_equipe_7[nome]=p

dic_elenco_8 = {}
dic_equipe_8 = {}
for nome in f8_elenco:
    p = (f8_elenco_recorrencia[nome]+1)/(total_elenco)
    dic_elenco_8[nome]=p
for nome in f8_equipe:
    p = (f8_equipe_recorrencia[nome]+1)/(total_equipe)
    dic_equipe_8[nome]=p
```

```python
# Listas com todos os nomes possíveis de atores/equipe.
membros_elenco_total = atores_elenco_3+atores_elenco_4+atores_elenco_5+atores_elenco_6+atores_elenco_7+atores_elenco_8
membros_equipe_total = membros_equipe_3+membros_equipe_4+membros_equipe_5+membros_equipe_6+membros_equipe_7+membros_equipe_8

# Adiciona os dicionários de elenco em uma lista.
lista_elenco = []
lista_elenco.append(dic_elenco_3)
lista_elenco.append(dic_elenco_4)
lista_elenco.append(dic_elenco_5)
lista_elenco.append(dic_elenco_6)
lista_elenco.append(dic_elenco_7)
lista_elenco.append(dic_elenco_8)

# Adiciona os dicionários de equipe em uma lista.
lista_equipe = []
lista_equipe.append(dic_equipe_3)
lista_equipe.append(dic_equipe_4)
lista_equipe.append(dic_equipe_5)
lista_equipe.append(dic_equipe_6)
lista_equipe.append(dic_equipe_7)
lista_equipe.append(dic_equipe_8)
```

```python
## Função que verifica em qual categoria é mais provável que cada ator apareça, e lhe atribui uma 'nota média'.

def calcula_maior_prob_elenco(nome):
    lista = []
    for dic in lista_elenco:
        for ator in dic:
            if ator == nome:
                lista.append(dic[nome])
    if (len(lista)>=1):
        nota_maior = max(lista)
        return nota_maior
    return 0

dic_elenco_media = {}
for nome in membros_elenco_total:
    x = calcula_maior_prob_elenco(nome)
    if nome in dic_elenco_3:
        if x == dic_elenco_3[nome]:
            dic_elenco_media[nome] = 3.5
    if nome in dic_elenco_4:
        if x == dic_elenco_4[nome]:
            dic_elenco_media[nome] = 4.5
    if nome in dic_elenco_5:
        if x == dic_elenco_5[nome]:
            dic_elenco_media[nome] = 5.5
    if nome in dic_elenco_6:
        if x == dic_elenco_6[nome]:
            dic_elenco_media[nome] = 6.5
    if nome in dic_elenco_7:
        if x == dic_elenco_7[nome]:
            dic_elenco_media[nome] = 7.5
    if nome in dic_elenco_8:
        if x == dic_elenco_8[nome]:
            dic_elenco_media[nome] = 8.5
```

```python
## Função que verifica em qual categoria é mais provável que cada ator apareça, e lhe atribui uma 'nota média'.

def calcula_maior_prob_equipe(nome):
    lista = []
    for dic in lista_equipe:
        for membro in dic:
            if membro == nome:
                lista.append(dic[membro])
    if (len(lista)>=1):
        nota_maior = max(lista)
        return nota_maior
    return 0

dic_equipe_media = {}
for nome in membros_equipe_total:
    x = calcula_maior_prob_equipe(nome)
    if nome in dic_equipe_3:
        if x == dic_equipe_3[nome]:
            dic_equipe_media[nome] = 3.5
    if nome in dic_equipe_4:
        if x == dic_equipe_4[nome]:
            dic_equipe_media[nome] = 4.5
    if nome in dic_equipe_5:
        if x == dic_equipe_5[nome]:
            dic_equipe_media[nome] = 5.5
    if nome in dic_equipe_6:
        if x == dic_equipe_6[nome]:
            dic_equipe_media[nome] = 6.5
    if nome in dic_equipe_7:
        if x == dic_equipe_7[nome]:
            dic_equipe_media[nome] = 7.5
    if nome in dic_equipe_8:
        if x == dic_equipe_8[nome]:
            dic_equipe_media[nome] = 9
```

```python
### E finalmente, a função que classifica de acordo com as médias de cada membro no elenco e na equipe.
nota = 0

def Predição_de_sucesso(lista):
    soma_elenco = 0
    soma1 = 0
    soma_equipe = 0
    soma2 = 0
    titulo = ''
    
    for membro in lista:
        for funcao in membro:
            if funcao == 'Ator' or funcao == 'Atriz':
                soma_elenco = soma_elenco
                soma1 = soma1
                if membro[funcao] in dic_elenco_media:
                    soma_elenco = soma_elenco + dic_elenco_media[membro[funcao]]
                    soma1 = soma1 + 1
            if funcao == 'Diretor' or funcao == 'Roteirista' or funcao == 'Produtor':
                soma = soma_equipe
                soma2 = soma2
                if membro[funcao] in dic_equipe_media:
                    soma_equipe = soma_equipe + dic_equipe_media[membro[funcao]]
                    soma2 = soma2 + 1

        try:
            titulo = membro['Título']
        except KeyError:
            pass
        
    nota_final = 0
    
    if soma1 > 0 and soma_elenco > 0:
        nota_elenco = soma_elenco/soma1
    else:
        nota_elenco = 0
        
    if soma2 > 0 and soma_equipe > 0:
        nota_equipe = soma_equipe/soma2
    
    else:
        nota_equipe = 0
    
        
    if nota_elenco != 0 and nota_equipe !=0: 
        # A equipe tem um peso de 65% na classificação, o elenco 35%. Um filme com atores fracos porém com direção
        # forte tem mais chances de ser bom que do um filme com bons atores, porém com direção fraca.
        nota_final = nota_elenco*0.35 + nota_equipe*0.65
   
    if nota_elenco == 0 and nota_equipe != 0:
        nota_final = nota_equipe
    
    if nota_elenco != 0 and nota_equipe == 0:
        nota_final = nota_elenco
    
    nota_final = round(nota_final,1)
    
    lista = []
    
    if nota_final <5:
        status = 'Que pena! Talvez o filme não seja muito bom...'
    if nota_final >= 5 and nota_final <= 6.9:
        status = 'Pelo visto, o filme é mediano.'
    if nota_final >=7:
        status = 'Uau! Esse filme provavelmente é muito bom!'
        
 #   if titulo != 0:
 #       return(print("A nota prevista do filme '",titulo,"' é",nota_final,'.',status))
    
 #   if titulo == 0:
 #       return(print('A nota prevista do filme é',nota_final,'.',status))
 #   return lista   

    lista.append(nota_final)
    
    return nota_final
```

```python
from tkinter import *
from PIL import ImageTk, Image
import os

import ctypes

myappid = 'Prophet Popcorn 2.0'
ctypes.windll.shell32.SetCurrentProcessExplicitAppUserModelID(myappid)

class pagina_inicial(object):
    def __init__(self,janela_inicial):
        self.instancia = janela_inicial 
        self.frame_pg_inicial = Frame(janela_inicial)
        
        self.label_bem_vindo = Label(janela_inicial, text = 'Bem-vindo ao Prophet Popcorn 2.0, seu previsor de sucessos da 7ª arte!',bg = 'red',fg='white',font=40)
        self.bt_iniciar = Button(self.frame_pg_inicial, text='COMEÇAR',bg = 'green', fg = 'white',font = 50,command=self.botao_comecar)
        self.label_bem_vindo.pack(pady=80)
        
        self.frame_pg_inicial.pack(side=TOP,pady=10)
        self.bt_iniciar.pack()
         
    def botao_comecar(self):
        self.instancia.destroy()
        self.inicia_instrucoes()
  
    def inicia_instrucoes(self):
        self.pg_instrucoes = Tk()
        self.pg_instrucoes.title('Prophet Popcorn 2.0')
        self.pg_instrucoes.wm_iconbitmap('icone_pipoca_de_oclinhos.ico')
        self.pg_instrucoes.geometry('600x378')
        self.pg_instrucoes.resizable(width=FALSE, height=FALSE)
        self.image = Image.open('cinema6.png')
        self.photo_image = ImageTk.PhotoImage(self.image)
        self.label = Label(self.pg_instrucoes, image = self.photo_image)
        self.label.place(x=0, y=0, relwidth=1, relheight=1)
        
        instrucoes(self.pg_instrucoes)
        
        self.pg_instrucoes.mainloop()
        
        
class instrucoes(object):
    def __init__(self,instrucoes):
        self.instancia = instrucoes
        self.frame_inst = Frame(instrucoes)
        
        self.frame_inst.pack(pady=30)
        
        self.subframe1 = Frame(self.frame_inst)
        self.subframe2 = Frame(self.frame_inst)
        self.subframe3 = Frame(self.frame_inst)
        self.subframe4 = Frame(self.frame_inst)
        self.subframe5 = Frame(self.frame_inst)
        self.subframe6 = Frame(self.frame_inst)
        self.subframe7 = Frame(self.frame_inst)
        self.subframe8 = Frame(self.frame_inst)
        self.subframe9 = Frame(self.frame_inst)
        self.subframe10 = Frame(self.frame_inst)
        self.subframe11 = Frame(self.frame_inst)
        self.subframe12 = Frame(self.frame_inst)
        self.subframe13 = Frame(self.frame_inst)
        
        self.subframe1.pack(side=TOP)
        self.subframe2.pack(side=TOP)
        self.subframe3.pack(side=TOP)
        self.subframe4.pack(side=TOP)
        self.subframe5.pack(side=TOP)
        self.subframe6.pack(side=TOP)
        self.subframe7.pack(side=TOP)
        self.subframe8.pack(side=TOP)
        self.subframe9.pack(side=TOP)
        self.subframe10.pack(side=TOP)
        self.subframe11.pack(side=TOP)
        self.subframe12.pack(side=TOP)
        self.subframe13.pack(side=TOP)
        
        self.label_inst1 = Label(self.subframe1,text='INSTRUÇÕES',font=40)
        self.label_inst2 = Label(self.subframe2,text='  O Prophet Popcorn é um programa que prediz a nota de um determinado filme')
        self.label_inst3 = Label(self.subframe3,text='baseado nas notas que os membros do elenco e da produção obtiveram em')
        self.label_inst4 = Label(self.subframe4,text='filmes anteriores. Por exemplo: se determinado ator participou de qua-')
        self.label_inst5 = Label(self.subframe5,text='tro filmes com nota 7,0 e dois de nota 5,0, é mais provável que o mes-')
        self.label_inst6 = Label(self.subframe6,text='mo participe de filmes de nota 7,0. Com base nisso, o programa predirá')
        self.label_inst7 = Label(self.subframe7,text='a nota de seus próximos filmes. Legal, não? Então vamos ver como utilizar:')
        self.label_inst8 = Label(self.subframe8,text=' ')
        self.label_inst9 = Label(self.subframe9,text='1 - Clique em "Criar item" para adicionar um novo membro no elenco/equipe')
        self.label_inst10 = Label(self.subframe10,text= '2 - Selecione a função do membro dentro do filme (Ator, Diretor, etc)')
        self.label_inst11 = Label(self.subframe11,text='3 - Clique em "Adicionar ao cálculo" para incluí-lo na lista de participantes')
        self.label_inst12 = Label(self.subframe12,text='4 - Quando tiver adicionado todos os integrantes desejados, clique em PREVEJA.')
        self.label_inst13 = Label(self.subframe13,text='ATENÇÃO: respeite as letras maiúsculas nos nomes. O nº máximo de nomes é 8.')
        
        self.label_inst1.pack()
        self.label_inst2.pack(side = LEFT)
        self.label_inst3.pack(side = LEFT)
        self.label_inst4.pack(side = LEFT)
        self.label_inst5.pack(side = LEFT)
        self.label_inst6.pack(side = LEFT)
        self.label_inst7.pack(side = LEFT)
        self.label_inst8.pack(side = LEFT)
        self.label_inst9.pack(side = LEFT)
        self.label_inst10.pack(side = LEFT)
        self.label_inst11.pack(side = LEFT)
        self.label_inst12.pack(side = LEFT)
        self.label_inst13.pack(side = LEFT)
        
        self.bt_vamola = Button(instrucoes,text= 'Entendi, vamos lá!', bg='green',fg='white',activeforeground = 'green',command=self.vamola)
        self.bt_vamola.pack()
        
    def vamola(self):
        self.instancia.destroy()
        self.janela2 = Tk()

        self.janela2.title('Prophet Popcorn 2.0')
        self.janela2.wm_iconbitmap('icone_pipoca_de_oclinhos.ico')
        self.janela2.geometry('600x378')
        self.janela2.resizable(width=FALSE, height=FALSE)
        self.image = Image.open('cinema6.png')
        self.photo_image = ImageTk.PhotoImage(self.image)
        self.label = Label(self.janela2, image = self.photo_image)
        self.label.place(x=0, y=0, relwidth=1, relheight=1)

        segunda_pagina(self.janela2)

        self.janela2.mainloop()


class segunda_pagina(object):
    def __init__(self,janela2):
        
        self.instancia = janela2
        self.frame0 = Frame(janela2)
        self.frame1 = Frame(janela2)
        self.frame2 = Frame(janela2)
        self.frame3 = Frame(janela2)
        self.frame4 = Frame()
        
        self.subframe0 = Frame(self.frame0,bd=2,bg='white')
        self.bt_inicio = Button(self.frame0, text = '< Voltar',bg='red',fg='white',activeforeground='red',command=self.voltar)
        self.bt_inicio.pack()
        
        self.bt_adicionar = Button(self.frame1,width=20,text='Criar item',command = self.adicionar_campo_de_entrada,bg='red',fg='white', activebackground='white', activeforeground='red')
        self.bt_adicionar.pack()
        
        self.bt_calcular = Button(self.frame3,width=20,text='PREVEJA',bg='green',fg='white',activeforeground='green',command=self.calcular)
        self.bt_calcular.pack(side=RIGHT)
        
        self.bt_ok = Button(self.frame3,width=20,text='Adicionar ao cálculo',bg='red',fg='white',activeforeground='red',command=self.bt_ok)
        self.bt_ok.pack(side=LEFT)
        
    
        self.frame0.pack(pady=0, padx=0, anchor=SW)
        self.frame1.pack(side=TOP,pady=1)
        self.frame2.pack()
        self.frame3.pack(side=BOTTOM,pady=30)
        
        self.lista = []
        
        self.titulo = 0
        
    def voltar(self):
        self.instancia.destroy()
        
        self.janela_inicial = Tk()
        self.janela_inicial.title('Prophet Popcorn 2.0')
        self.janela_inicial.wm_iconbitmap('icone_pipoca_de_oclinhos.ico')
        self.janela_inicial.geometry('600x378')
        self.janela_inicial.resizable(width=FALSE, height=FALSE)
        self.image = Image.open('cinema5.png')
        self.photo_image = ImageTk.PhotoImage(self.image)
        self.label = Label(self.janela_inicial, image = self.photo_image)
        self.label.place(x=0, y=0, relwidth=1, relheight=1)

        pagina_inicial(self.janela_inicial)
        self.janela_inicial.mainloop()
    
    
    def bt_ok(self):
        self.label['text'] = 'Adicionado'
        self.label['fg'] = 'green'
        dic = {}
        if self.variavel.get() == 'Ator/Atriz':
            dic['Ator'] = self.campo.get()
            self.lista.append(dic)
        if self.variavel.get() == 'Título':
            dic['Título'] = self.campo.get()
            self.titulo = dic['Título']
            self.lista.append(dic)
        if self.variavel.get() == 'Diretor(a)':
            dic['Diretor'] = self.campo.get()
            self.lista.append(dic)
        if self.variavel.get() == 'Produtor(a)':
            dic['Produtor'] = self.campo.get()
            self.lista.append(dic)
            
        
    def fecha_janela(self):
        self.janela3.destroy()
        self.instancia.destroy()
        self.novo_teste()

    def adicionar_campo_de_entrada(self):
        subframe = Frame(self.frame2)
        subframe.pack(side=TOP,pady=0)
        self.campo = Entry(subframe)
        self.label = Label(subframe,text='Não adicionado',fg='red')
        
        self.variavel = StringVar(subframe)
        self.variavel.set("Selecione")
        self.menu = OptionMenu(subframe,self.variavel,"Título","Ator/Atriz","Diretor(a)",'Roteirista','Produtor(a)')
        
        self.label.pack(side=RIGHT)
        self.campo.pack(side=LEFT)
        self.menu.pack(side=RIGHT)
        
        
    def calcular(self):
        self.inicia_janela3()
        
    def inicia_janela3(self):
        self.janela_3()
          
    def janela_3(self):
        
        resultado = Predição_de_sucesso(self.lista)
        self.janela3 = Toplevel()
        self.janela3.title('Prophet Popcorn 2.0')
        self.janela3.wm_iconbitmap('icone_pipoca_de_oclinhos.ico')
        self.janela3.geometry('600x378')
        self.janela3.resizable(width=FALSE, height=FALSE)
        self.image = Image.open('cinema6.png')
        self.photo_image = ImageTk.PhotoImage(self.image)
        self.label = Label(self.janela3, image = self.photo_image)
        self.label.place(x=0, y=0, relwidth=1, relheight=1)
        
        self.subframezeras = Frame(self.janela3)
        
        self.botao_voltar = Button(self.janela3, text='Realizar nova predição', bg='red', fg='white', activeforeground='red',command=self.fecha_janela)
        
        if self.titulo != 0:
        
            self.label_nota = Label(self.subframezeras,text = 'A nota prevista do filme')
            self.label_titulo = Label(self.subframezeras,text=self.titulo)
            self.label_eh = Label(self.subframezeras,text='é:')
            self.label_resultado = Label(self.janela3)
            self.label_status = Label(self.janela3)
            
            self.label_titulo.config(font=(14))
            self.label_eh.config(font=(14))
        
        if self.titulo == 0:
            self.label_nota = Label(self.janela3,text = 'A nota prevista do filme é')
            self.label_resultado = Label(self.janela3)
            self.label_status = Label(self.janela3)
        
        if resultado >= 7:
            self.label_resultado['bg'] = 'green'
            self.label_resultado['fg'] = 'white'
            self.label_status['text'] = 'Uau! Esse filme provavelmente é muito bom!'
            
        if resultado >= 5 and resultado < 7:
            self.label_resultado['bg'] = 'blue'
            self.label_resultado['fg'] = 'white'
            self.label_status['text'] = 'Pelo visto, o filme é mediano.'
            
        if resultado < 5:
            self.label_resultado['bg'] = 'red'
            self.label_resultado['fg'] = 'white'
            self.label_status['text'] = 'Que pena! Talvez o filme não seja muito bom...'
        
        self.label_resultado['text'] = str(resultado)
        
        self.label_nota.config(font=(14))
        self.label_resultado.config(font=(14))
        self.label_status.config(font=(14))
        
        if self.titulo == 0:
            self.label_nota.pack(side=TOP,pady=60)
            self.label_resultado.pack(side=TOP)
            self.label_status.pack(side=TOP,pady=60)
            self.botao_voltar.pack(side=TOP,padx=10)
    
        if self.titulo != 0:
            self.label_nota.pack(side=LEFT)
            self.label_titulo.pack(side=LEFT)
            self.label_eh.pack(side=LEFT)
            self.subframezeras.pack(side=TOP, pady=60)
            self.label_resultado.pack(side=TOP)
            self.label_status.pack(side=TOP,pady=60)
            self.botao_voltar.pack(side=TOP,padx=10)
    
    def novo_teste(self):
        self.janela = Tk()
        self.janela.title('Prophet Popcorn 2.0')
        self.janela.wm_iconbitmap('icone_pipoca_de_oclinhos.ico')
        self.janela.geometry('600x378')
        self.janela.resizable(width=FALSE, height=FALSE)
        self.image2 = Image.open('cinema6.png')
        self.photo_image2 = ImageTk.PhotoImage(self.image2)
        self.label2 = Label(self.janela, image = self.photo_image2)
        self.label2.place(x=0, y=0, relwidth=1, relheight=1)       
        
        segunda_pagina(self.janela)
        self.janela.mainloop()

janela_inicial = Tk()
janela_inicial.title('Prophet Popcorn 2.0')
janela_inicial.wm_iconbitmap('icone_pipoca_de_oclinhos.ico')
janela_inicial.geometry('600x378')
janela_inicial.resizable(width=FALSE, height=FALSE)
image = Image.open('cinema5.png')
photo_image = ImageTk.PhotoImage(image)
label = Label(janela_inicial, image = photo_image)
label.place(x=0, y=0, relwidth=1, relheight=1)

pagina_inicial(janela_inicial)
janela_inicial.mainloop()
```
